```python
amenities_df = pd.get_dummies(df1['amenities'].apply(pd.Series).stack()).groupby(level=0).sum()
df2 = pd.concat([df1, amenities_df], axis=1)
```

```python
host_df = df2.explode('host', ignore_index=True)

#  Normalize the 'host' JSON data into individual columns
host_df = pd.json_normalize(df2['host'])
```

```python
host_df.drop(['host_url','host_about','host_picture_url','host_verifications', 'host_response_time', 'host_thumbnail_url','host_neighbourhood','host_response_rate'], axis=1, inplace=True)
```

```python
df3 = pd.concat([df2, host_df], axis=1)
```

```python
df3
```

```python
address_df = df3.explode('address', ignore_index=True)

# Normalize the 'address' JSON data into individual columns

address_df = pd.json_normalize(df3['address'])
df4 = pd.concat([df3, address_df], axis=1)
```

```python
df4
```

```python
availability_df = df4.explode('availability', ignore_index=True)
# Normalize the 'availability' JSON data into individual columns
availability_df = pd.json_normalize(df4['availability'])
df5 = pd.concat([df4, availability_df], axis=1)
```

```python
df['bathrooms'] = df['bathrooms'].astype(str).astype(float)
```

```python
# Convert 'Decimal128' to 'float' for 'security_deposit' and 'cleaning_fee' columns
df['security_deposit'] = df['security_deposit'].astype(str).astype(float)
df['cleaning_fee'] = df['cleaning_fee'].astype(str).astype(float)
```

```python
# Convert the 'price' column to a numeric data type (e.g., float64)
df['price'] = df['price'].astype(str).astype('float64')
```

```python
# Impute missing values in 'bedrooms' column with the median
df['bedrooms'].fillna(df['bedrooms'].median(), inplace=True)
```

```python
# Impute missing values in 'bathrooms' column with the mean
df['bathrooms'].fillna(df['bathrooms'].mean(), inplace=True)
```

```python
# Impute missing values in 'bed_type' column with the mode
df['bed_type'].fillna(df['bed_type'].mode()[0], inplace=True)
```

```python
# Calculate the median for 'security_deposit' and 'cleaning_fee'
median_security_deposit = df['security_deposit'].median()
median_cleaning_fee = df['cleaning_fee'].median()
```

```python
# Impute missing values in 'security_deposit' and 'cleaning_fee' columns with the median
df['security_deposit'].fillna(median_security_deposit, inplace=True)
df['cleaning_fee'].fillna(median_cleaning_fee, inplace=True)
```

```python
# Impute missing values in 'first_review' and 'last_review' columns with earliest and latest dates, respectively
earliest_date = df['first_review'].min()
latest_date = df['last_review'].max()
df['first_review'].fillna(earliest_date, inplace=True)
df['last_review'].fillna(latest_date, inplace=True)
```

```python
# Calculate the mean or median for 'beds'
mean_beds = df['beds'].mean()
median_beds = df['beds'].median()

# Impute missing values in 'beds' column with the mean or median
df['beds'].fillna(mean_beds, inplace=True)  # or df['beds'].fillna(median_beds, inplace=True)
```

# Bivariant analysis

```python
# Price Analysis by Property Type
average_price_by_property_type = df.groupby('property_type')['price'].mean().reset_index()
average_price_by_property_type = average_price_by_property_type.sort_values(by='price', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x='property_type', y='price', data=average_price_by_property_type)
plt.xticks(rotation=90)
plt.title('Average Price by Property Type')
plt.xlabel('Property Type')
plt.ylabel('Average Price')
plt.show()
```

**Maps To Analyse The Spread of Listings**

```python
# Create a folium map centered at the mean of the coordinates
mean_lat = df6['location.coordinates'].apply(lambda x: x[1]).mean()
mean_lon = df6['location.coordinates'].apply(lambda x: x[0]).mean()
map_airbnb = folium.Map(location=[mean_lat, mean_lon], zoom_start=10)

# Extract the latitude and longitude pairs from the DataFrame and convert to a list of lists
locations = df6['location.coordinates'].apply(lambda x: [x[1], x[0]]).tolist()

# Plot the heatmap based on the locations
HeatMap(locations).add_to(map_airbnb)

# Display the map
map_airbnb
```
